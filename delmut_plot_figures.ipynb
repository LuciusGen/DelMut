## Oryza vs Pisum predictions

#### Libraries and paths

```r
library(plyr)
library(ggplot2)
options(repr.plot.width=5, repr.plot.height=4)
path_predictions = 'classifiers/all/'
```

#### Data and plotting

```r
d = read.table(paste0(path_rf_accuracies,'Random_Forest_wout_weights_accuracies.tsv'), header = TRUE)

x = count(as.data.frame(d), vars=c('Acc_Pisum','Acc_Oryza'))

x1 = max(x[,'Acc_Oryza'])
x2 = max(x[x[,'Acc_Oryza'] == x1,'Acc_Pisum'])

c(x1, x2)
colnames(x) <- c('Pisum', 'Oryza', 'Frequency')

 
# Color and shape depend on factor (categorical variable)
ggplot(x, aes(x=Pisum, y=Oryza, color=Frequency, size=Frequency)) +
    geom_point(alpha=0.6) + geom_point(aes(x=x2, y=x1), size=1, colour="red")

```
